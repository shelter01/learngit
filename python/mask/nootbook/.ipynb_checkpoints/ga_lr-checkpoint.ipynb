```python
import pandas as pd
import numpy as np
import random
import math
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
point = pd.read_csv('d:/sjw/experimental/regression/point.csv',header=None)#停车数据
center = pd.read_csv('d:/sjw/experimental/regression/center.csv',header=None)#停车点数据
```

```python
#转化成列表
point = point.values.tolist()
center = center.values.tolist()
```

```python
#生成初始停车顺序
V = len(point)
P = len(center)
pop = 10000
p_max = round(V/P/0.9)
l = []
lr_x = []#线性回归矩阵
lr_y = []
for i in range(V):#初始化1-n
    l.append(i)
#画图相关
gen = 100#隔多少代画图
min_x = []
min_ga = []
ga_lr = []
plt.ion()   
```

```python
for i in range(pop):#训练集
    print(i)
    
    relist = l
    random.shuffle(relist)#乱序停车顺序
    
    mindis = 0
    lr_x.append([])
    p_sum = []
    for x in range(P):
        p_sum.append(0)
    for x in range(V):
        lr_x[i].append(0)
    
    for j in range(V):#遍历停车队列
        a = 0
        for k in range(P):#遍历带停车辆到每个停车点的距离
            t_dis = np.sqrt(np.sum(np.square(np.array(point[relist[j]])-np.array(center[k]))))
            if k == 0:#初始化理想最近距离和停车点编号
                v_min = t_dis
                v_num = k
            
            if t_dis < v_min:#更新得到最近的理想距离和编号
                v_min = t_dis
                v_num = k
                
            if p_sum[k] < p_max:#实际有停车位时
                if a == 0:
                    dis = t_dis
                    t_num = k
                    a = 1
                if t_dis < dis:
                    dis = t_dis
                    t_num = k
                    
        if t_num == v_num:#判断是否停到了最近的停车位
            lr_x[i][relist[j]] = 1
        else:
            lr_x[i][relist[j]] = 0
                
        mindis = mindis + dis
        p_sum[t_num] = p_sum[t_num] + 1
        
    lr_y.append(mindis)
    
        
    if (i%gen == 0) and (i != 0):
        #ga的搜索最小值
        if i == gen:
            min_ga.append(mindis)
            ga_lr.append(mindis)
        elif mindis < min_ga[-1]:
            min_ga.append(mindis)
            ga_lr.append(mindis)
        else:
            min_ga.append(min_ga[-1])
            ga_lr.append(min_ga[-1])

        #训练lr
        y = lr_y
        y = np.reshape(y,newshape=(i+1,1))
        # 调用模型
        lr = LinearRegression()
        # 训练模型
        lr.fit(lr_x,y)
        lastlist = np.argsort(lr.coef_[0],axis=0)
    
        #画图相关参数
        relist = lastlist
        mindis = 0
        p_sum = []
        for x in range(P):
            p_sum.append(0)
    
        for j in range(V):#遍历停车队列
            a = 0
            for k in range(P):#遍历带停车辆到每个停车点的距离
                t_dis = np.sqrt(np.sum(np.square(np.array(point[relist[j]])-np.array(center[k]))))
                if k == 0:#初始化理想最近距离和停车点编号
                    v_min = t_dis
                    v_num = k
            
                if t_dis < v_min:#更新得到最近的理想距离和编号
                    v_min = t_dis
                    v_num = k
                
                if p_sum[k] < p_max:#实际有停车位时
                    if a == 0:
                        dis = t_dis
                        t_num = k
                        a = 1
                    if t_dis < dis:
                        dis = t_dis
                        t_num = k
                
            mindis = mindis + dis
            p_sum[t_num] = p_sum[t_num] + 1

        #整合
        if mindis < ga_lr[-1]:
            ga_lr[-1] = mindis
        
        min_x.append(i)
        
        #画图
        plt.clf()                  # 清除之前画的图
        plt.plot(min_x,ga_lr)
        plt.plot(min_x,min_ga,'--')# 画出当前 ax 列表和 ay 列表中的值的图形 
        plt.pause(0.1) # 关闭画图的窗口
        plt.ioff()
```
