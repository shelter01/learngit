```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline

x = [[123,210,120],[640,641,760],[980,450,241]]
y = [21332, 20162, 19138]
y = np.reshape(y,newshape=(3,1))
# 调用模型
lr = LinearRegression()
# 训练模型
lr.fit(x,y)
# 计算R平方
print (lr.score(x,y))
print (lr.coef_)
# 计算y_hat
y_hat = lr.predict(x)
# 打印出图
plt.plot(x, y_hat)
```
